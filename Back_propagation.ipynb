#### Back Propagation

```python
x = torch.tensor(4.0,requires_grad = True)
```

```python
y = x** 2

print(x,y)
```

```python
y.backward()
```

```python
y
```

```python
print(x.grad)
```

```python
lst = [[2.,3.,1.],[4.,6.,5.],[6.,7.,8.]]
       
torch_input = torch.tensor(lst,requires_grad= True)
```

```python
y = torch_input**3+torch_input**2
y
```

```python
z = y.sum()
z
```

```python
z.backward
```

```python
torch_input.grad
```
